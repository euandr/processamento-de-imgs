## Preparando a base de dados

Esta etapa não é o foco do nosso exercício. Sendo assim, só irei reproduzir o passo a passo do link passado anteriormente.

```python

import pandas as pd
import numpy as np

#Lendo a base de dados
df = pd.read_csv('/content/train.csv')

print("Tabela Original")
df.head()
# display(df)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

df = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
df = df.dropna()
df['Gender'] = df['Sex'].map({'female': 0, 'male':1}).astype(int)
df['Port'] = df['Embarked'].map({'C':1, 'S':2, 'Q':3}).astype(int)
df = df.drop(['Sex', 'Embarked'], axis=1)
cols = df.columns.tolist()
cols = [cols[1]] + cols[0:1] + cols[2:]
df = df[cols]


print("Tabela depois de processada")
df.head()
```

Por fim, geramos nossa base de treino e teste no formato para o ScikitLearn

```python
dataset = {
    'data': df.values[0:,2:],
    'target': df.values[0:,0]
}
```

```python
X = dataset['data']
y = dataset['target']
```

# Editando a partir daqui


```python
# separar a base em dados de trino e teste
from sklearn.model_selection import train_test_split
xtreino,xteste,ytreino,yteste = train_test_split(X,y,test_size=0.1)

#criando modelo
from sklearn.neighbors   import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
#treinando
knn.fit(xtreino,ytreino)

#aplicando conhecimento
previsao=knn.predict(xteste)

acertou=0
for questao,resposta in zip(previsao,yteste):
  if questao==resposta:
    acertou+=1
print(f"acertou: {acertou}, de {len(previsao)}" )
print(f'errou: {len(previsao)-acertou}')
```

```python
#verificando resultando
from sklearn.metrics import accuracy_score
accuracy_score(yteste,previsao)
```
